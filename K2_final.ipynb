```python
import sympy as sp
from sympy import *
from sympy import MatMul
from sympy import symbols, Matrix, eye, simplify, Integer
```

```python
D=Matrix([[Rational(0.854373111798292)  ,-Rational(0.518782521594128)] , [ Rational(0.518794363700548) ,  Rational(0.848187547437653)]])
# best found example 
N(D.norm(ord=2))
# the norm of D is <1
```

```python
a00=Integer(1); a01=-D[1,1]; a10=-D[0,0] ; a11=D[0,0]*D[1,1]-D[1,0]*D[0,1];
# here we form the coefficients of p(z)
# p(z) = det(I-D[[z1,0],[0,z2]])
# q = bar p


z1, z2 = symbols('z1 z2')
p = a11*z1*z2 + a10*z1 + a01*z2 + a00
q = a00*z1*z2 + a01*z1 + a10*z2 + a11
```

```python
# 1st method of obtaining the expansion of q/p: using sympy.expand


def bivariate_series_tensor(q, p, z1, z2, deg=12):
  
    f = sp.simplify(q/p)

    
    # Expand in both variables
    series_z1 = sp.series(f, z1, 0, deg+1).removeO()
    series_full = sp.series(series_z1, z2, 0, deg+1).removeO()
    f_exp = sp.expand(series_full)
   
    
    # Create tensor
    A = sp.zeros(deg+1, deg+1)
    
    # Get all coefficients using SymPyâ€™s Poly
    poly = sp.Poly(f_exp, z1, z2)
    for (i, j), coeff in poly.terms():
        #if i + j <= deg:
        A[i, j] = coeff
    return A


A = bivariate_series_tensor(q, p, z1, z2, deg=12)
# the matrix of coefficients c_alpha, alpha=(alpha1,alpha2)
```

```python
# verification that equation (34) holds
r=Rational(.3177)
R=Matrix([r**0, r**1, r**2, r**3, r**4, r**5, r**6, r**7, r**8, r**9,r**10,r**11,r**12] )
g=(R.T)*abs(A)*R
N(g)
```

```python

# 2nd method of obtaining the expansion of q/p:  solving a system of linear equations in exact artithmetics
# Obviously, this is the same as the first method, the difference is only in computer implementation 
# this code can be easily converted to, e.g., matlab to derive furhter numerical computations or veriffications

n=13;  subeye=Matrix([zeros(1,n),eye(n,n-1).T]);
AD=a00*eye(n)+a01*subeye;
AS=a10*eye(n)+a11*subeye;

from sympy.physics.quantum import TensorProduct
U=   TensorProduct(eye(n),AD)+TensorProduct(subeye,AS); 
b=Matrix([a11,a10,zeros(n-2,1),a01,a00,zeros(n-2,1),zeros(n**2-2*n,1)]);
c=U.LUsolve(b)
from sympy import reshape
C=c.reshape(n,n);




```
